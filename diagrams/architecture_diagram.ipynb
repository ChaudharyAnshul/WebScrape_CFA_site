```python
from diagrams import Diagram, Cluster, Edge
from diagrams.aws.storage import SimpleStorageServiceS3
from diagrams import Cluster, Diagram
from diagrams.programming.language import Python
from urllib.request import urlretrieve
from diagrams.custom import Custom
from diagrams.digitalocean.storage import Folder
from diagrams.oci.connectivity import DNS
from diagrams.elastic.beats import Filebeat
```

```python
graph_attr = {
  "width": "1000",
  "height": "1000"
}

with Diagram("\n\nArchitecture Diagram \n\n\nWorkflow : Structured Database and Text Extraction \nSystem for Finance Professional Development Resources", show=False, graph_attr=graph_attr, direction="LR") as diag:

  cfa_website = DNS("CFA Websites")

  pdf_file = Filebeat("PDF Files")

  with Cluster("\nWeb Scrapping"):
    py_web_scrapping = Python("\nExtract\nCFA Data")

  with Cluster("\nPDF Extraction"):
    py_pdf_extraction = Python("\nExtract\nPDF Text")

  with Cluster("\nFile Upload"):
    py_upload_pdf = Python("\nUpload\nPDF Text")

  with Cluster("\nDatabase Upload"):
    py_upload_data = Python("\nUpload\nStructured Data")

  with Cluster("\nCloud Storage"):
    with Cluster("\nStructured Data"):
      snowflake_db_url = "https://pbs.twimg.com/media/EDT-rsuWwAAmhav.png"
      snowflake_db_icon = "elastic.png"
      urlretrieve(snowflake_db_url, snowflake_db_icon)
      snowflake_db = Custom("Snowflake\nData Warehouse", snowflake_db_icon, edge_label=" ", x_pos="1", y_pos="1")

    with Cluster("\nFile Storage"):
      s3_bucket = SimpleStorageServiceS3("\nS3 Bucket")

  with Cluster("\nLocal Storage"):
    local_storage = Folder("\nTemporary\nStorage")

  cfa_website \
    >> Edge(color="darkgreen") \
    >> py_web_scrapping \
    >> Edge(color="darkgreen") \
    >> local_storage \
    >> Edge(color="darkgreen") \
    >> py_upload_data \
    >> Edge(color="darkgreen") \
    >> snowflake_db

  pdf_file \
    >> Edge(color="darkorange") \
    >> py_pdf_extraction \
    >> Edge(color="darkorange") \
    >> local_storage \
    >> Edge(color="darkorange") \
    >> py_upload_pdf \
    >> Edge(color="darkorange") \
    >> snowflake_db

  py_upload_pdf \
    >> Edge(color="darkorange") \
    >> s3_bucket

diag
```
